# Histograms

```python
def histogram_creator (x, title = "", xlab = "", ylab = "",  color = "blue", edgecolor = "black", grid = False, save = False, **kwargs):
    plt.xlabel(f"{xlab}")
    plt.ylabel(f"{ylab}")
    title_font= {"family": "sans-serif", "color" : "black", "size" : 16, "weight" : "bold"} # create a different font for the title of the plot (https://how.dev/answers/how-to-set-font-properties-for-title-and-labels-in-matplotlib, https://matplotlib.org/stable/users/explain/text/text_props.html)
    plt.title (f"{title}", fontdict= title_font)
    plt.hist(x, edgecolor = edgecolor, color = color, **kwargs)
    if grid == True:
        plt.grid(alpha = 0.2)
    if save == True:
        plt.savefig(f"{title}.png")
    plt.show()
```

```python
# Indexing columns in Pandas DataFrame (https://pandas.pydata.org/docs/user_guide/indexing.html)
sl = iris.data.iloc[:, 0] # sepal length
sw = iris.data.iloc[:, 1] # sepal width
pl = iris.data.iloc[:, 2] # petal length
pw = iris.data.iloc[:, 3] # petal width
```

```python
histogram_creator(sl, title = "Sepal Length Histogram", xlab = "Sepal Length", ylab= "Frequency", grid = True);
```

```python
histogram_creator(sw, title = "Sepal Width Histogram", xlab = "Sepal Width", ylab = "Frequency", color = "green", grid = True)
```

```python
histogram_creator(pl, title = "Petal Length Histogram", xlab = "Petal Length", ylab = "Frequency", color = "red", grid = True)
```

```python
histogram_creator(pw, title = "Petal Width Histogram", xlab = "Petal Width", ylab = "Frequency", color = "orange", grid = True)
```

# Scatterplots

```python
def scatterplot_creator (x, y, xlab = "", ylab = "", c = "red", title = "", grid = False, save = False, fitline = False, **kwargs):
    plt.scatter(x, y , c = c, **kwargs)
    plt.xlabel(f"{xlab}")
    plt.ylabel(f"{ylab}")
    title_font= {"family": "sans-serif", "color" : "black", "size" : 16, "weight" : "bold"}
    plt.title(f"{title}", fontdict= title_font)
    if grid == True:
        plt.grid(alpha = 0.2)
    if save == True:
        plt.savefig(f"{title}.png")
    if fitline == True:
        m, c = np.polyfit(x, y, deg = 1)
        plt.plot(x, m*x + c, color = "red")
        plt.legend (["Data", "Line of Best Fit"])
    plt.show()
```

```python
species = iris["target"] 
```

```python
scatterplot_creator (sl, sw, xlab= "Sepal Length (cm)", ylab= "Sepal Width (cm)", title= "Sepal Length vs Sepal Width", c = species)
```

```python
scatterplot_creator(sl, pl, xlab= "Sepal Length (cm)", ylab= "Petal Length (cm)", title= "Sepal Length vs Petal Length", c= species)
```

```python
scatterplot_creator(sl, pw, xlab= "Sepal Length (cm)", ylab= "Petal Width (cm)", title= "Sepal Length vs Sepal Width", c= species)
```

```python
scatterplot_creator(pl, sw, xlab= "Petal Length (cm)", ylab= "Sepal Width (cm)", title= "Petal Length vs Sepal Width", c= species)
```

```python
scatterplot_creator(pw, sw, xlab= "Petal Width", ylab= "Sepal Width (cm)", title= "Petal Width vs Sepal Width", c= species)
```

```python
scatterplot_creator(pl, pw, xlab= "Petal Length (cm)", ylab= "Petal Width (cm)", title= "Petal Length vs Petal Width", c= species)
```
