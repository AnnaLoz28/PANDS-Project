# Features Analysis

```python
# Indexing columns in Pandas DataFrame (https://pandas.pydata.org/docs/user_guide/indexing.html)
sl = iris.data.iloc[:, 0] # sepal length
sw = iris.data.iloc[:, 1] # sepal width
pl = iris.data.iloc[:, 2] # petal length
pw = iris.data.iloc[:, 3] # petal width
```

```python
# create a function that performs the features analysis and returns the results
def feature_analysis (x):
    mean = np.mean(x) # mean
    median = np.median(x) # median
    std = np.std(x) # standard deviation
    min = np.min(x) # minimum value
    max = np.max(x) # maximum value 
    return(f"Mean: {mean:.2f}.\nMedian: {median:.2f}.\nStandard deviation: {std:.2f}.\nMinimum value: {min:.2f}.\nMaximum value: {max:.2f}.")
```

Store and print out the results of feature analysis:

```python

sl_analysis = feature_analysis(sl) # analysis of sepal length
print(sl_analysis)
```

```python
sw_analysis = feature_analysis(sw) # analysis of sepal width
print(sw_analysis)
```

```python
pl_analysis = feature_analysis(pl) # analysis of petal length
print(pl_analysis)
```

```python
pw_analysis = feature_analysis(pw) # analysis of petal width
print(pw_analysis)
```

Summarize the analysis of all the features in a DataFrame:

```python
# creation of a DataFrame with features summary  (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html)
features_summary= pd.DataFrame({
    "Mean" : iris["data"].mean(),
    "Median" : iris["data"].median(),
    "Standard Deviation" : iris["data"].std(),
    "Minimum" : iris["data"].min(),
    "Maximum" : iris["data"].max()
})
```

```python
# print out the DataFrame
print(features_summary)
```

Create a text file to store the output of iris features analysis:

```python
# create a text file with features analysis data
with open ("Iris Dataset Variables Analysis.txt", "w") as f:
    f.write(f"IRIS SEPAL LENGTH ANALYSIS:\n{sl_analysis}\n\n\n")
    f.write(f"IRIS SEPAL WIDTH ANALYSIS:\n{sw_analysis}\n\n\n")
    f.write(f"IRIS PETAL LENGTH ANALYSIS:\n{pl_analysis}\n\n\n")
    f.write(f"IRIS PETAL WIDTH ANALYSIS:\n{pw_analysis}\n\n\n")
    f.write(f"FEATURES SUMMARY:\n{features_summary}")
```

# Histograms

For the creation of histogram plots Matplotlib will be used.  
Matplotlib is a very useful Python library that can be used to create different types of plots and visualize data.  
https://matplotlib.org/

A histogram is a plot showing the frequency distribution (on the y-axes) of a variable (on the x-axis).  
It shows the distribution of the variable values within a given interval.

https://www.w3schools.com/python/matplotlib_histograms.asp

```python
# define a function that create histogram plots
def histogram_creator (x, title = "", xlab = "", ylab = "",  color = "blue", edgecolor = "black", grid = False, save = False, **kwargs): # define the arguments of the function
    plt.xlabel(f"{xlab}") # define the x axes label
    plt.ylabel(f"{ylab}") # define the y axes label
    title_font= {"family": "sans-serif", "color" : "black", "size" : 16, "weight" : "bold"} # create a different font for the title of the plot (https://how.dev/answers/how-to-set-font-properties-for-title-and-labels-in-matplotlib, https://matplotlib.org/stable/users/explain/text/text_props.html)
    plt.title (f"{title}", fontdict= title_font) # define the title of the plot
    plt.hist(x, edgecolor = edgecolor, color = color, **kwargs) # create the plot
    if grid == True: # define how the grid will be visualized by default
        plt.grid(alpha = 0.2) # grid transparency set on 80%
    if save == True: # define how the image of the plot will be saved by default
        plt.savefig(f"{title}.png")
    plt.show()
```

```python
# create the histogram plot of Sepal Length (cm)
histogram_creator(sl, title = "Sepal Length Histogram", xlab = "Sepal Length (cm)", ylab= "Frequency", grid = True);
```

The histogram plot of sepal length shows that most flowers have a sepal length around 5.5 - 6.0 cm, while few have very short and very long sepals.  
The distribution range is broad, but we can see 3 peaks. If these 3 peaks are related to different species of *Iris* flowers, this feature may be helpful in discriminating them.

```python
# create the histogram plot of Sepal Width (cm)
histogram_creator(sw, title = "Sepal Width Histogram", xlab = "Sepal Width (cm)", ylab = "Frequency", color = "green", grid = True)
```

The histogram plot of sepal width shows that many flowers have a sepal width between 2.5 and 3.3 cm, while few have very narrow or broad sepals. 

```python
# create the histogram plot of Petal Length (cm)
histogram_creator(pl, title = "Petal Length Histogram", xlab = "Petal Length (cm)", ylab = "Frequency", color = "red", grid = True)
```

The histogram plot of petal length shows 2 peaks, with a bimodal distribution type.  
There is a great separation between short and long petals, which can help classify flowers if these values are related to the *Iris* species.

```python
# create the histogram plot of Petal Width (cm)
histogram_creator(pw, title = "Petal Width Histogram", xlab = "Petal Width (cm)", ylab = "Frequency", color = "orange", grid = True)
```

The histogram plot of petal width shows a bimodal or trimodal distribution.  
This feature can be informative and can be used to classify *Iris* flowers if it's related to the species.

# Scatterplots

A scatterplot displays individual data points on a two-dimensional plane. Each point represents the values of two variables, with one plotted on the x-axis and the other on the y-axis.  
Scatterplots help visualize relationships, correlations, and patterns between variables.

https://en.wikipedia.org/wiki/Scatter_plot

```python
# define a function that creates scatterplots
def scatterplot_creator (x, y, xlab = "", ylab = "", c = "red", title = "", grid = False, save = False, fitline = False, **kwargs): # define the arguments of the function
    plt.scatter(x, y , c = c, **kwargs) # create the plot (https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html)
    plt.xlabel(f"{xlab}") # define the x axes label
    plt.ylabel(f"{ylab}") # define the y axes label
    title_font= {"family": "sans-serif", "color" : "black", "size" : 16, "weight" : "bold"} # use a custom font for the title
    plt.title(f"{title}", fontdict= title_font) # define the title of the plot
    if grid == True: # define how grid will appear by default
        plt.grid(alpha = 0.2) # grid transparency set on 80%
    if save == True: # define how images of plots will be saved by default
        plt.savefig(f"{title}.png")
    if fitline == True: # calculate the parameters of the regression line to examine the linear relationship between variables (https://numpy.org/doc/stable/reference/generated/numpy.polyfit.html, https://www.geeksforgeeks.org/line-of-best-fit/ )
        m, c = np.polyfit(x, y, deg = 1) # slope (m), intercept (c)
        plt.plot(x, m*x + c, color = "red") # plot of the regression line
        plt.legend (["Data", "Line of Best Fit"])
    plt.show()
```

## Boxplots

A boxplot is a useful tool that graphically demonstrates how data is distributed using quartiles. It allows us to visualize skewness, detect outliers, assess central tendency (such as the median), and compare different datasets more easily.  
A boxplot shows important information, such as:

- Minimum 

- First quartile (Q1), 25th percentile

- Median (Q2), the middle value of a dataset

- Third quartile (Q3), 75th percentile

- Maximum

- The Interquartile range (IQR), which is Q3 - Q1

- Outliers

Each quartile corresponds to 25% of data. Whiskers are useful for visualizing the minimum and maximum values within a certain range, while any ouliers (extremely large or small values) are shown as individual data points outside the whiskers.  
https://en.wikipedia.org/wiki/Box_plot  
https://www.khanacademy.org/math/statistics-probability/summarizing-quantitative-data/box-whisker-plots/a/box-plot-review  


As we have seen, there are three classes of *Iris* flowers: *setosa*, *versicolor* and *virginica*, identified as 0, 1, and 2, respectively. 

```python
# define the species based on target number
setosa = iris["data"][iris["target"] == 0]
versicolor = iris["data"][iris["target"] == 1]
virginica = iris["data"][iris["target"] == 2]
```

```python
# define a function that creates boxplots
def boxplots_creator (x, title = "", xlab = "", ylab= "", ticklabs = "", grid = False, save = False, **kwargs):
    fig, ax = plt.subplots() # option to create multiple plots on the same figure (https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html)
    ax.boxplot(x, patch_artist= True) # create boxplot filled with color
    title_font= {"family": "sans-serif", "color" : "black", "size" : 16, "weight" : "bold"} # create a custom font for the title
    ax.set_ylabel (f"{ylab}", fontsize = 12) # set the y axes label
    ax.set_xlabel(f"{xlab}", fontsize = 12) # set the x axes label
    ax.set_xticklabels(ticklabs) # create labels for each tick mark on the x axes
    ax.set_title (f"{title}", fontdict = title_font) # set a title 
    if grid == True: # define how grid will be visualized by deafult
        ax.grid(axis = "y", alpha = 0.5, linestyle = "--") # grid transparency set on 50%, grid made with dashed line, located on the y axes
    if save == True:  # define how the image will be savedby default
        plt.savefig(f"{title}.png")
    plt.show()
    
```

Exploring petal lengths of the different species:

```python
# define petal lengths of the different species
setosa_pl = setosa["petal length (cm)"]
versicolor_pl = versicolor["petal length (cm)"]
virginica_pl = virginica ["petal length (cm)"]
```

```python
# create the boxplot for Iris setosa petal length
boxplots_creator(setosa_pl, title= "Iris Setosa Petal Length", xlab= "Iris setosa", ylab="Petal Length (cm)", grid = True)
```

```python
# create the boxplot for Iris versicolor petal length
boxplots_creator(versicolor_pl, title= "Iris Versicolor Petal Length", xlab= "Iris versicolor", ylab="Petal Length (cm)", grid = True )
```

```python
# create the boxplot for Iris virginica petal length
boxplots_creator(virginica_pl, title= "Iris Virginica Petal Length", xlab= "Iris virginica", ylab="Petal Length (cm)", grid = True )
```

```python
# visualize all the boxplots in one figure
boxplots_creator([setosa_pl, versicolor_pl, virginica_pl], ticklabs= ["setosa", "versicolor", "virginica"], xlab= "Species", ylab= "Petal Length (cm)", title = "Petal Lengths of All Species", grid = True)
```

The boxplots for petal length across the 3 *Iris* species show a strong visual separation.  
*Iris setosa* shows the the shortest petal length with little variation, while *versicolor* and *virginica* have longer and more varied petal lengths.  
The petal length appears to be a highly informative feature for species classification, as the 3 species data almost do not overlap, especially with *setosa* boxplot being very far apart from the others.
